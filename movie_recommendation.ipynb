```python
def tag(genre):
  lis=[]
  for i in ast.literal_eval(genre):
    lis.append(i["name"])
  return lis
```

```python
import ast
movies['genres']=movies['genres'].apply(tag)
```

```python
movies['keywords']=movies['keywords'].apply(tag)
```

```python
# Extract the names of the first three cast members from the JSON-like string and return them as a list.

def tag1(cast):
  l=[]
  s=1
  for i in ast.literal_eval(cast):
    if s!=4:
       l.append(i["name"])
       s+=1
    else:
      break
  return l
```

```python
movies['cast']=movies['cast'].apply(tag1)
```

```python
movies.head()
```

```python
def director(crew):
  lis=[]
  for i in ast.literal_eval(crew):
    if i["job"]=='Director':
     lis.append(i['name'])
     break
  return lis
```

```python
movies['crew']=movies['crew'].apply(director)
```

```python
movies.head(1)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(1)
```

```python
def year(yr):
  l=[]
  l.append(int(yr))
  return l
```

```python
def year2(yr):
  l=[]
  l.append(str(yr))
  return l
```

```python
movies['year']=movies['year'].apply(year)
movies['year']=movies['year'].apply(year2)
```

```python
movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['tags']=movies['genres']+movies['cast']+movies['keywords']+movies['crew']+movies['overview']+movies['year']
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
# Stemming of words into one common word
def stem(words):
  l=[]
  for i in words.split():
    l.append(ps.stem(i))

  return " ".join(l)
```

```python
movies['tags']=movies['tags'].apply(stem)
```

```python
# Create a CountVectorizer object with specified parameters
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=6000,stop_words='english')
```

```python
vec=cv.fit_transform(movies['tags']).toarray()
vec.shape
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cs=cosine_similarity(vec)
cs.shape # Get the shape of the cosine similarity matrix
```
